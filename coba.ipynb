```python
# qq plot
import numpy as np 
import pylab 
import scipy.stats as stats
import pandas as pd

df = pd.read_excel('nilai.xls')
data_points = df['nilai uts']
# nilai_uts = data_points.to_numpy()
stats.probplot(data_points, dist="norm", plot=pylab)
pylab.show()
```

```python
# histogram
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_excel('nilai.xls')
data_points = df['nilai uts']
plt.hist(data_points, bins=10)
plt.show()
```
