```python
# qq plot
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_excel('nilai.xls')
nilai_uts = df['nilai uts']
nilai_uas = df['nilai uas']
nilai_tugas = df['nilai tugas']

nsample = 100
np.random.seed(7654321)

fig = plt.figure()
ax = fig.add_subplot(111)
x = stats.t.rvs(3, size=nsample)
stats.probplot(nilai_uts, plot=plt)
stats.probplot(nilai_tugas, plot=plt)
stats.probplot(nilai_uas, plot=plt)

ax.get_lines()[0].set_markerfacecolor('C0')
ax.get_lines()[1].set_markerfacecolor('C1')
ax.get_lines()[2].set_markerfacecolor('C2')

ax.get_lines()[1].remove()
ax.get_lines()[2].remove()

plt.show()
```

```python
# histogram
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_excel('nilai.xls')
nilai_uts = df['nilai uts']
nilai_uas = df['nilai uas']
nilai_tugas = df['nilai tugas']

n_bins = 20

fig, (ax0) = plt.subplots(nrows=1, ncols=1)

colors = ['red', 'tan', 'lime']
ax0.hist([nilai_uts,nilai_tugas,nilai_uas], n_bins, density=True, histtype='bar', color=colors, label=['nilai uts','nilai tugas','nilai uas'])
ax0.legend(prop={'size': 10})
ax0.set_title('bars with legend')
```

```python
# Scatter Plot
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_excel('nilai.xls')
nilai_uts = df['nilai uts']
nilai_uas = df['nilai uas']
nilai_tugas = df['nilai tugas']
nomor_mahasiswa = df['nomor mahasiswa']

fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.scatter(nilai_uts, nomor_mahasiswa, s=10, c='b', marker="s", label='nilai uts')
ax1.scatter(nilai_uas, nomor_mahasiswa, s=10, c='r', marker="o", label='nilai uas')
ax1.scatter(nilai_tugas, nomor_mahasiswa, s=10, c='g', marker="v", label='nilai tugas')
plt.legend(loc='upper left');
plt.show()
```
